```python
## Import Libraries
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import random
import os

os.makedirs("../images", exist_ok=True)
```

## API Used

OpenWeatherMap API is used to extract real-time weather data such as temperature and humidity.

Since real-time data does not change frequently within seconds, slight variations are introduced to simulate trends.

## API Key & Setup

```python
API_KEY = "6f08d197085e92809d33dd1436352419"

city = "Mumbai"

url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
```

## Data Extraction + Simulation

```python
records = []

for i in range(10):
    response = requests.get(url)
    data = response.json()

    temp = data["main"]["temp"]
    humidity = data["main"]["humidity"]

    temp = temp + random.uniform(-1, 1)
    humidity = humidity + random.randint(-3, 3)

    records.append({
        "Temperature": temp,
        "Humidity": humidity
    })

df = pd.DataFrame(records)

df
```

```python
# Data Cleaning
```

```python
df.isnull().sum()

df.drop_duplicates(inplace=True)
```

```python
# Save Dataset

df.to_csv("../data/weather_data.csv", index=False)
```

## VISUALIZATIONS

```python
# Temperature Trend

plt.figure(figsize=(8,5))

plt.plot(df["Temperature"], marker="o")

plt.title("Temperature Trend")
plt.xlabel("Time Index")
plt.ylabel("Temperature (°C)")

plt.tight_layout()
plt.savefig("../images/temperature_trend.png")

plt.show()
```

```python
# Humidity Trend

plt.figure(figsize=(8,5))

plt.bar(range(len(df)), df["Humidity"])

plt.title("Humidity Trend")
plt.xlabel("Time Index")
plt.ylabel("Humidity")

plt.tight_layout()
plt.savefig("../images/humidity_trend.png")

plt.show()
```

## Interpretation

- Temperature shows slight variation over time
- Humidity fluctuates based on atmospheric conditions
- Data demonstrates how real-time APIs can be used for trend analysis
